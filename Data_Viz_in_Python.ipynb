```
plt.scatter(data['sales'], data['TV'], color='green')
plt.title('Sales and Cost of TV ads', fontsize=14)
plt.xlabel('Sales', fontsize=14)
plt.ylabel('Cost of TV ads', fontsize=14)
plt.grid(True)
plt.show()
```

```
plt.scatter(data['sales'], data['radio'], color='blue')
plt.title('Sales and Cost of radio ads', fontsize=14)
plt.xlabel('Sales', fontsize=14)
plt.ylabel('Cost of radio ads', fontsize=14)
plt.grid(True)
plt.show()
```

```
plt.scatter(data['sales'], data['newspaper'], color='red')
plt.title('Sales and Cost of TV newspaper', fontsize=14)
plt.xlabel('Sales', fontsize=14)
plt.ylabel('Cost of newspaper ads', fontsize=14)
plt.grid(True)
plt.show()
```
